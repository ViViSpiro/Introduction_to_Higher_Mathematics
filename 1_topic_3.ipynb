```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-5, 5, 210)
y1 = 3 * np.cos(x - 2) + 5
y2 = 2 * np.cos(x - 1) + 4
y3 = 9 * np.cos(x - 6) + 2
plt.plot(x, y1, color="red")
plt.plot(x, y2, color="green")
plt.plot(x, y3, color="blue")
plt.xlabel("x")
plt.ylabel("y")
plt.grid(color="lightgray", linestyle="--")
plt.show()
```
