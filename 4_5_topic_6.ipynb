```python
from pylab import *
from mpl_toolkits.mplot3d import Axes3D

fig = figure()
ax = Axes3D(fig, auto_add_to_figure=False)
fig.add_axes(ax)
x = np.arange(-50, 50, 1)
y = np.arange(-50, 50, 1)
x, y = np.meshgrid(x, y)


def Q(x, y, z):
    return (x ** 2 + y ** 2 + z ** 2)


z_1 = -1 + x + 2 * y
y_1 = 0.5 - 0.5 * x + 0.5 * z_1
z_2 = 6 - 4 * x + 2.5 * y
y_2 = - 2.4 + 1.6 * x + 0.4 * z_2
ax.plot_surface(x, y, Q(x, y_1, z_1), color="green")
ax.plot_surface(x, y, Q(x, y_2, z_2), color="red")
plt.xlabel("x")
plt.ylabel("y")
plt.grid(True)
show()
A = np.array([[1, 2, -1], [8, -5, 2]])
B = np.array([1, 12])
print(np.linalg.lstsq(A, B, rcond=None))
```
