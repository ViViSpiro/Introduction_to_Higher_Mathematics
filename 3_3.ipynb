```python
import matplotlib.pyplot as plt
import numpy as np
import decimal

xmin = -20
xmax = 20
dx = 0.1

xlist = np.around(np.arange(xmin, xmax, dx), decimals=4)
ylist = 1 / xlist

plt.plot(xlist, ylist)
plt.title("Hyperbole")
plt.grid(color="lightgray", linestyle="--")
plt.show()
```

```python
import matplotlib.pyplot as plt

x_1 = []
x_2 = []
y_1 = []
y_2 = []
k = 1.0

for i in range(1000):
    if i != 0:
        x_ = i / 1000
        x_1.append(x_)
        x_2.append(- x_)
        y_1.append(k / x_)
        y_2.append(- k / x_)
plt.plot(x_1, y_1)
plt.plot(x_2, y_2)
plt.ylabel("y")
plt.xlabel("x")
plt.title("Hyperbole")
plt.grid(color="lightgray", linestyle="--")
plt.axis([-10, 10, -10, 10])
plt.show()
```
