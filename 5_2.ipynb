```python
from pylab import *
from mpl_toolkits.mplot3d import Axes3D

fig = figure()
ax = Axes3D(fig, auto_add_to_figure=False)
fig.add_axes(ax)
x = np.arange(-500, 500, 1)
y = np.arange(-500, 500, 1)
x, y = np.meshgrid(x, y)
z_1 = np.cos(x ** 2) + np.sin(y ** 2)
z_2 = - x ** 2 - y ** 2
ax.plot_surface(x, y, z_1, color="green")
ax.plot_surface(x, y, z_2, color="red")
show()
```
