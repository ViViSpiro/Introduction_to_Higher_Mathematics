```python
import numpy as np
from scipy.optimize import fsolve
import matplotlib.pyplot as plt

x = np.linspace(-2, 3, 201)
plt.plot(x, (np.exp(x) + x - 1) / x)
plt.plot(x, x ** 2 - 1)
plt.xlabel("x")
plt.ylabel("y")
plt.ylim(-2, 8)
plt.grid(True)
plt.show()


def equations(p):
    x, y = p
    print(x)
    print(y)
    return y - x ** 2 + 1, np.exp(x) + x - x * y - 1


result_1 = fsolve(equations, (-1, 1))
result_2 = fsolve(equations, (3, 6))
print(result_1)
print(result_2)
```
