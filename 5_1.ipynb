```python
from pylab import *
from mpl_toolkits.mplot3d import Axes3D


def draw_parallel_planes(A=0, B=0, C=1, D=0, z_offset=0):
    fig = figure()
    ax = Axes3D(fig, auto_add_to_figure=False)
    fig.add_axes(ax)
    x = np.arange(-50, 50, 1)
    y = np.arange(-50, 50, 1)
    x, y = np.meshgrid(x, y)
    z_1 = (A * x + B * y + D) / C
    z_2 = (A * x + B * y + D) / C + z_offset
    ax.plot_wireframe(x, y, z_1, color="green")
    ax.plot_wireframe(x, y, z_2, color="blue")
    ax.scatter(0, 0, 0, "z", 20, "red")
    show()


draw_parallel_planes(10, 12, z_offset=1000)
```
